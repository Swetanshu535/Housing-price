```python
missing_values = df.isnull().sum()
print(missing_values)
```

```python
unique_categories = df['ocean_proximity'].unique()
print(unique_categories)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(8,5))
sns.histplot(df["median_house_value"], bins=30, kde=True, color='blue')


plt.xlabel("Median House Value")
plt.ylabel("Frequency")
plt.title("Distribution of Median House Value")


plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x=df["median_income"], y=df["median_house_value"], alpha=0.5, color='blue')


plt.xlabel("Median Income")
plt.ylabel("Median House Value")
plt.title("Relationship between Median Income and Median House Value")


plt.show()
```
